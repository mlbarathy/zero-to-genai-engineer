## Section 1: Variables, Strings, and Lists

### The 3 building blocks of every GenAI program

**Analogy:** Think of a variable as a labeled box. You put something inside and give it a name. Whenever you need it, you just use the name.

In GenAI:
- Everything starts as **text** (a string)
- The model breaks text into **pieces called tokens** (stored in a list)
- Each token eventually becomes a **number** (so the model can do math on it)

This section covers the tools for all three.

````python
# ── VARIABLES AND STRINGS ──────────────────────────────────────────────

# A variable stores a value — you name it, then use the name later
sentence = "The cat sat on the mat"

# print() shows the value on screen
print(sentence)           # → The cat sat on the mat
print(type(sentence))     # → <class 'str'>  (str = string = text)

# Strings have built-in tools (called methods) — accessed with a dot
print(sentence.lower())   # → the cat sat on the mat  (all lowercase)
print(sentence.upper())   # → THE CAT SAT ON THE MAT  (all uppercase)
print(len(sentence))      # → 22  (number of characters including spaces)

# The most important string method for GenAI:
# .split() breaks a sentence into a list of words
words = sentence.split()
print(words)              # → ['The', 'cat', 'sat', 'on', 'the', 'mat']
````

````python
# ── LISTS ──────────────────────────────────────────────────────────────
# A list holds multiple items in order — items go inside square brackets []
# In GenAI, tokens (word pieces) are stored in lists

tokens = ['The', 'cat', 'sat', 'on', 'the', 'mat']

# Access items by position — counting starts at 0, not 1
print(tokens[0])    # → The    (first item, index 0)
print(tokens[1])    # → cat    (second item, index 1)
print(tokens[-1])   # → mat    (last item — negative index counts from end)

# Get a range of items (called a slice)
print(tokens[1:4])  # → ['cat', 'sat', 'on']  (items at index 1, 2, 3)

# How many items?
print(len(tokens))  # → 6

# Add an item to the end
tokens.append('!')
print(tokens)       # → ['The', 'cat', 'sat', 'on', 'the', 'mat', '!']

# Check if something is in a list
print('cat' in tokens)   # → True
print('dog' in tokens)   # → False
````

## Section 2: Loops and Functions

### How GenAI processes every word, one at a time

**Analogy for loops:** Imagine a factory conveyor belt. Every item on the belt goes through the same process. A loop does the same thing — it runs the same code for every item in a list.

**Analogy for functions:** A function is a named recipe. You write the steps once, give it a name, and reuse it wherever you need it — without rewriting the steps every time.

In GenAI code, you'll see loops and functions **constantly** — every tokenizer, every embedding, every pipeline uses them.

````python
# ── FOR LOOPS ──────────────────────────────────────────────────────────
# A for loop runs the same code for each item in a list
# Syntax: for <item_name> in <list>:
#             (indented code runs once per item)

tokens = ['the', 'cat', 'sat', 'on', 'the', 'mat']

# Print each token on its own line
print("--- All tokens ---")
for token in tokens:
    print(token)   # this line runs 6 times, once per token

print()

# Do something with each token — count its characters
print("--- Token lengths ---")
for token in tokens:
    length = len(token)                    # count characters in this token
    print(f"{token} → {length} characters")  # print token and its length

print()

# Collect results into a new list using .append()
lengths = []                  # start with an empty list
for token in tokens:
    lengths.append(len(token))  # add each length to the list

print("Lengths list:", lengths)  # → [3, 3, 3, 2, 3, 3]
````

````python
# ── FUNCTIONS ──────────────────────────────────────────────────────────
# A function is a named block of code you can reuse
# def <name>(<inputs>):
#     (indented code)
#     return <output>

# Example: a function that cleans a sentence for processing
def clean_text(text):
    """Takes a sentence and returns it cleaned (lowercase, no punctuation)."""
    text = text.lower()                    # make everything lowercase
    text = text.replace('.', '')           # remove periods
    text = text.replace(',', '')           # remove commas
    text = text.replace('!', '')           # remove exclamation marks
    return text                            # send the result back to the caller

# Call the function by name — pass input in parentheses
result = clean_text("Hello, World! This is GenAI.")
print(result)   # → hello world this is genai

print()

# Functions can call other functions — this is how pipelines are built
def tokenize(text):
    """Cleans text and splits it into tokens."""
    cleaned = clean_text(text)   # call the function we wrote above
    tokens = cleaned.split()     # split into words
    return tokens

sentence = "The model learns, one token at a time!"
tokens = tokenize(sentence)
print(tokens)   # → ['the', 'model', 'learns', 'one', 'token', 'at', 'a', 'time']
````

## Section 3: Dictionaries and f-strings

### The tools LLMs use to map words to numbers

**Analogy for dictionaries:** A dictionary is like a phone book. You look up a name (key) and get a number (value). In GenAI, we look up a word (key) and get its ID number (value). This is literally how a tokenizer vocabulary works.

**f-strings** are how you build prompts dynamically — inserting variables into text. You'll use them every single session.

````python
# ── DICTIONARIES ───────────────────────────────────────────────────────
# A dictionary stores key → value pairs
# Syntax: {key: value, key: value, ...}

# Example: a simple vocabulary (word → ID number)
# This is exactly what a real tokenizer does internally
vocab = {
    'the':   0,
    'cat':   1,
    'sat':   2,
    'on':    3,
    'mat':   4
}

# Look up a word's ID
print(vocab['cat'])    # → 1
print(vocab['the'])    # → 0

# Add a new word
vocab['dog'] = 5
print(vocab)           # → {'the': 0, 'cat': 1, 'sat': 2, 'on': 3, 'mat': 4, 'dog': 5}

# Check if a word is in the vocab
print('cat' in vocab)  # → True
print('lion' in vocab) # → False

# Loop over a dictionary
print("\n--- Vocabulary ---")
for word, id_num in vocab.items():      # .items() gives you key AND value
    print(f"{word} → ID {id_num}")
````

````python
# ── F-STRINGS ──────────────────────────────────────────────────────────
# An f-string lets you insert variables directly into text
# Prefix the string with f, then put variables inside {}

name = "Mohamed"
topic = "transformers"

# Without f-string (messy)
message1 = "Hello " + name + ", today we learn " + topic
print(message1)

# With f-string (clean)
message2 = f"Hello {name}, today we learn {topic}"
print(message2)

print()

# In GenAI, f-strings build prompts dynamically
def build_prompt(user_question, context):
    """Builds a prompt by inserting user input into a template."""
    prompt = f"""You are a helpful assistant.

Context: {context}

Question: {user_question}

Answer:"""
    return prompt

question = "What is a transformer?"
context = "Transformers are neural networks that use attention mechanisms."

final_prompt = build_prompt(question, context)
print(final_prompt)
````
